CASE STUDY 1: DEPARTMENT STORE DATASET

```python
nl4dv_instance1 = NL4DV(data_url="https://raw.githubusercontent.com/frog-land/Chat2VIS_Streamlit/main/department_store.csv")
```

```python
case1 = nl4dv_instance1.analyze_query("What is the highest price of product, grouped by product type? Show a bar chart, and display by the names in desc.")
```

```python
try:
    display({'application/vnd.vega.v5+json': case1['visList'][0]['vlSpec']}, raw=True)
except Exception as e: print(e)
```
